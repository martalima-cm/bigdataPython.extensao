# Saber a diferença entre os dados

```python
sns.pairplot(df, hue = 'race/ethnicity')
```

**Correlação positiva:** A inclinação geral dos pontos sugere uma correlação positiva entre as notas das diferentes matérias. Isso significa que, em geral, os alunos que obtiveram notas mais altas em uma matéria também tendem a obter notas mais altas nas outras matérias.

**Variabilidade:** A dispersão dos pontos indica que a relação entre as notas não é perfeita. Existem alunos que se destacam em uma matéria e não em outra, e vice-versa.

## Desempenho por Grupo:
**Média**: A posição das curvas de densidade ao longo do eixo indica a média das notas para cada grupo em cada matéria. Por exemplo, se a curva de um grupo estiver mais à direita, significa que, em média, os alunos desse grupo obtiveram notas mais altas.
**Dispersão:** A largura das curvas de densidade indica a variabilidade das notas dentro de cada grupo. Curvas mais estreitas indicam menor variabilidade, ou seja, as notas dos alunos desse grupo estão mais concentradas em torno da média.

**Sobreposição:** A sobreposição das curvas indica que há uma certa sobreposição nas distribuições de notas entre os diferentes grupos. Isso significa que não há uma separação clara entre os grupos em termos de desempenho.

### Observações
**Outliers**: Pontos isolados podem representar outliers, ou seja, valores que se desviam significativamente dos demais.

**Interações**: A análise visual sugere que pode haver interações entre as variáveis. Por exemplo, a relação entre as notas de matemática e leitura pode ser diferente para os diferentes grupos raciais ou étnicos.
